```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
#strandradization - feature scaling techniqe
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train_scaler=scaler.fit_transform(X_train)
```

```python
X_test_scaler=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## eLbow method - to select k value
wcss=[]
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,init="k-means++")
  kmeans.fit(X_train_scaler)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#determin k
# plot elbow curve
plt.plot(range(1,11),wcss)
plt.title("Elbow curve")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
kmeans.fit_predict(X_train_scaler)
```

```python
plt.scatter(X_train_scaler[:,0],X_train_scaler[:,1],c=kmeans.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhoutee_coefficients=[]
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,init="k-means++")
  kmeans.fit(X_train_scaler)
  score=silhouette_score(X_train_scaler,kmeans.labels_)
  silhoutee_coefficients.append(score)
```

```python
silhoutee_coefficients
```

```python
plt.plot(range(2,11),silhoutee_coefficients)
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.title("Silhouette Coefficient vs. Number of Clusters")
plt.show()
```
